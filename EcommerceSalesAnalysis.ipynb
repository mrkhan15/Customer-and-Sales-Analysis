```python
#Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#deleting unrelated or blank columns
df.drop(['New','PendingS'], axis=1, inplace=True)
df.info()
```

```python
# Checking for null values
pd.isnull(df)
```

```python
#Checking total null values
pd.isnull(df).sum()
```

```python
df.shape
```

```python
#Removing null values
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.info()
```

```python
#Changing data type for ship-postal-code column
df['ship-postal-code'] = df['ship-postal-code'].astype('int')
```

```python
#Changing data type for Date column
df['Date'] =pd.to_datetime(df['Date'])
```

```python
#Validating data type change
df['ship-postal-code'].dtype
```

```python
#Validating data type change for Date column
df['Date'].dtype
```
